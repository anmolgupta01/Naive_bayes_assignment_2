# Q1. A company conducted a survey of its employees and found that 70% of the employees use the company's health insurance plan, while 40% of the employees who use the plan are smokers. What is the probability that an employee is a smoker given that he/she uses the health insurance plan ?

To find the probability that an employee is a smoker given that he/she uses the health insurance plan, we can use Bayes' theorem. Let's define the events:

- \( A \): Employee uses the health insurance plan.
- \( B \): Employee is a smoker.

The probability that an employee uses the health insurance plan is denoted as P(A) = 0.70 (70% of employees).

The probability that an employee is a smoker is denoted as P(B) . However, this information is not directly given. We can use the information provided to calculate P(B|A) , the probability that an employee is a smoker given that he/she uses the health insurance plan.

The conditional probability P(B|A)  is given by:

 P(B|A) = P(A intersection B) / P(A)

Here:
-  P(A intersection B)  is the probability that an employee both uses the health insurance plan and is a smoker. This is given as 40% of employees who use the plan, so  P(A intersection B = 0.40 \times 0.70 \).
-  P(A)  is the probability that an employee uses the health insurance plan, given as 70%, so  P(A) = 0.70 

Now, we can calculate  P(B|A) 

 P(B|A) = P(A intersection B)*P(A) = {0.40 *0.70}/{0.70}

The  P(A)  terms cancel out:

 P(B|A) = 0.40

Therefore, the probability that an employee is a smoker given that he/she uses the health insurance plan is 40%.

# Q2. What is the difference between Bernoulli Naive Bayes and Multinomial Naive Bayes?

Both Bernoulli Naive Bayes and Multinomial Naive Bayes are variants of the Naive Bayes algorithm, which is a probabilistic classifier based on Bayes' theorem. The main difference between them lies in the type of data they are designed to handle:

1. **Bernoulli Naive Bayes:**
   - **Data Type:** Bernoulli Naive Bayes is suitable for binary data, where each feature represents a binary value (0 or 1).
   - **Use Case:** It is commonly used for text classification tasks, where the presence or absence of words in a document is considered.
   - **Example:** Document classification where the features are the presence or absence of specific words in the document.

2. **Multinomial Naive Bayes:**
   - **Data Type:** Multinomial Naive Bayes is designed for count-based data, where features represent the frequency with which certain events occur (typically integer counts).
   - **Use Case:** It is commonly used for document classification tasks, where the features are word frequencies or term frequencies in a document.
   - **Example:** Document classification where the features are the counts of words in the document.

In summary, Bernoulli Naive Bayes is suitable for binary data, while Multinomial Naive Bayes is more appropriate for count-based data. The choice between them depends on the nature of your features and the type of data you are working with.

# Q3. How does Bernoulli Naive Bayes handle missing values?

In the context of Bernoulli Naive Bayes, the handling of missing values depends on how the algorithm is implemented or the specific library/tool you are using. However, in general, Bernoulli Naive Bayes assumes that the absence of a feature is as important as its presence. Therefore, missing values are often treated as if the feature is not present.

Here's a simplified explanation of how Bernoulli Naive Bayes typically handles missing values:

1. **Assumption of Absence:** In a Bernoulli Naive Bayes model, each feature is treated as a binary variable (present or absent). When a feature is missing, it is usually assumed to be absent.

2. **Zero Probability for Missing Features:** In the context of probability calculations, if a feature is missing, its contribution to the likelihood calculation is considered as zero. This means that the absence of the feature is treated as if it has a probability of zero, and it contributes accordingly to the overall probability calculation.

3. **Impact on Class Probability:** When estimating class probabilities for a given instance, the absence of a feature (missing value) is taken into account along with the presence of other features. The absence of a feature can affect the likelihood of a particular class.

It's important to note that the treatment of missing values in Bernoulli Naive Bayes may vary depending on the specific implementation or library you are using. Some implementations may allow you to explicitly handle missing values, while others may follow the assumptions mentioned above. Always refer to the documentation of the specific implementation or library for detailed information on how missing values are treated.

# Q4. Can Gaussian Naive Bayes be used for multi-class classification?

Yes, Gaussian Naive Bayes can be used for multi-class classification. The Gaussian Naive Bayes algorithm is an extension of the Naive Bayes algorithm, specifically designed for continuous data. It assumes that the features follow a Gaussian (normal) distribution.

For multi-class classification, the algorithm can be extended to handle more than two classes. The general approach is to apply the Gaussian Naive Bayes algorithm independently for each class and then assign the class with the highest probability as the predicted class.

Here's a basic overview of how Gaussian Naive Bayes can be used for multi-class classification:

1. **Parameter Estimation:** For each class, the algorithm estimates the mean and variance of each feature assuming a Gaussian distribution.

2. **Class Probability:** Calculate the prior probability of each class, i.e., the probability of each class occurring without considering any features.

3. **Likelihood:** For each feature in each class, calculate the likelihood using the Gaussian probability density function.

4. **Posterior Probability:** Combine the prior probability and the likelihood for each class to compute the posterior probability for each class.

5. **Prediction:** Assign the class with the highest posterior probability as the predicted class.

This process is repeated for each class, and the algorithm can handle scenarios where there are more than two classes. Each class is modeled independently, and the assumption of independence between features within a class, though simplified, is often effective in practice.

Keep in mind that the choice of Naive Bayes variant (Gaussian, Bernoulli, or Multinomial) depends on the nature of your data. If your features are continuous and follow a Gaussian distribution, then Gaussian Naive Bayes is a suitable choice for multi-class classification.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_csv('spambase.data')
```

```python
dict_1={}

dict_1=dict(df.corr()['1'])
```

```python
list_features=[]
for key,values in dict_1.items():
    if abs(values)<0.2:
        list_features.append(key)
```

```python
#lets drop the features which have coorelation less than <0.2
df=df.drop(list_features,axis=1)
```

```python
df
```

```python
y=df['1']
X=df.drop(['1'],axis=1)
```

# Naive bayes classification

```python
# Bernoulli Naive Bayes
bernoulli_nb = BernoulliNB()
bernoulli_scores = cross_val_score(bernoulli_nb, X, y, cv=10)
```

```python
# Multinomial Naive Bayes
multinomial_nb = MultinomialNB()
multinomial_scores = cross_val_score(multinomial_nb, X, y, cv=10)
```

```python
# Gaussian Naive Bayes
gaussian_nb = GaussianNB()
gaussian_scores = cross_val_score(gaussian_nb, X, y, cv=10)
```

```python
# Function to print performance metrics
def print_metrics(name, scores):
    print(f"\n{name} Naive Bayes Performance Metrics:")
    print(f"Accuracy: {scores['accuracy'].mean()}")
    print(f"Precision: {scores['precision'].mean()}")
    print(f"Recall: {scores['recall'].mean()}")
    print(f"F1 Score: {scores['f1'].mean()}")
```

```python
# Calculate performance metrics
performance_scores = {
    "bernoulli": {
        "accuracy": bernoulli_scores.mean(),
        "precision": precision_score(y, bernoulli_nb.fit(X, y).predict(X)),
        "recall": recall_score(y, bernoulli_nb.fit(X, y).predict(X)),
        "f1": f1_score(y, bernoulli_nb.fit(X, y).predict(X))
    },
    "multinomial": {
        "accuracy": multinomial_scores.mean(),
        "precision": precision_score(y, multinomial_nb.fit(X, y).predict(X)),
        "recall": recall_score(y, multinomial_nb.fit(X, y).predict(X)),
        "f1": f1_score(y, multinomial_nb.fit(X, y).predict(X))
    },
    "gaussian": {
        "accuracy": gaussian_scores.mean(),
        "precision": precision_score(y, gaussian_nb.fit(X, y).predict(X)),
        "recall": recall_score(y, gaussian_nb.fit(X, y).predict(X)),
        "f1": f1_score(y, gaussian_nb.fit(X, y).predict(X))
    }
}
```

```python
# Print performance metrics
print_metrics("Bernoulli", performance_scores["bernoulli"])
print_metrics("Multinomial", performance_scores["multinomial"])
print_metrics("Gaussian", performance_scores["gaussian"])
```
